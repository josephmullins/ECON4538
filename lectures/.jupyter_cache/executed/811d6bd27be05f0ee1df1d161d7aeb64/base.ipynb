```julia
#| echo: false

using Plots, LaTeXStrings

# Parameters for the example
y_L = 1.0
y_H = 3.0
 = 0.7  # probability of low income for low type

# Calculate key values
E_y = 0.5 * y_L + 0.5 * y_H  # average income
E_log_y = 0.5 * log(y_L) + 0.5 * log(y_H)  # E[u(y)] in laissez-faire
C_star = E_y  # optimal consumption (perfect insurance)
log_C_star = log(C_star)  # u(C*)

# Certainty equivalents for each type in laissez-faire
# For type L: *log(y_L) + (1-)*log(y_H) = log(c_L)
# For type H: (1-)*log(y_L) + *log(y_H) = log(c_H)
c_bar_L = exp( * log(y_L) + (1-) * log(y_H))
c_bar_H = exp((1-) * log(y_L) +  * log(y_H))
C_bar = 0.5 * c_bar_L + 0.5 * c_bar_H  # average certainty equivalent
log_C_bar = log(C_bar)  # u(C)

# Create the plot
y_range = range(0.5, 3.5, length=100)
plot(y_range, log.(y_range),
     label="",
     linewidth=3,
     color=:blue,
     xlabel="Consumption",
     ylabel="Utility (log)",
     title="Decomposition: Insurance vs Redistribution",
     legend=:bottomright,
     size=(800, 500))

# Mark the income levels
scatter!([y_L, y_H], [log(y_L), log(y_H)],
         markersize=8,
         color=:red,
         label="Income realizations")

# E[log(y)] - expected utility in laissez-faire
hline!([E_log_y],
       linestyle=:dash,
       color=:red,
       linewidth=2,
       label=L"E[\log(y)]")

# log(C) - utility of average certainty equivalent
hline!([log_C_bar],
       linestyle=:dash,
       color=:orange,
       linewidth=2,
       label=L"\log(\bar{C})")

# log(C*) - utility under perfect insurance
hline!([log_C_star],
       linestyle=:dash,
       color=:green,
       linewidth=2,
       label=L"\log(C^*)")

# Add vertical line at E[y]
vline!([E_y],
       linestyle=:dot,
       color=:gray,
       linewidth=1.5,
       label=L"E[y]")

# Add annotations for the decomposition
annotate!(3.3, E_log_y + 0.05, text("Laissez-faire", :red, :left, 9))
annotate!(3.3, log_C_bar + 0.05, text("+ Redistribution", :orange, :left, 9))
annotate!(3.3, log_C_star + 0.05, text("+ Insurance", :green, :left, 9))

# Add arrows showing the gaps
arrow_x = 0.7
annotate!(arrow_x, (E_log_y + log_C_bar)/2,
          text("Redistribution\ngain", :orange, :center, 8))
annotate!(arrow_x, (log_C_bar + log_C_star)/2,
          text("Insurance\ngain", :green, :center, 8))

plot!()
```
